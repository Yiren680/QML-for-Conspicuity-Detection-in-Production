- Basic Concepts

    - Quantum bits (qubits)

        - Notion of qubits

            Classical computers represent information using bits, which are binary values that can be either 0 or 1. This can be associated with a physical system existing in one of two distinct states.

            The power of quantum computers lies in their unique way of representing and manipulating information. They use special units called quantum bits, or qubits. Unlike classical bits, which are typically represented by electronic charges or magnetic states corresponding to 0 or 1, qubits can exist in a superposition of both states simultaneously. Qubits also have many physical implementations. It can be carried by a single atom, or a single electron, or a single photon (a particle of light). Or a qubit can be carried by a more complicated system, like a very cold superconducting electrical circuit in which many electrons are moving.
            
            For example, quantum information can be encoded in the polarization, phase, or path of photons, making them useful for qubits, especially in quantum communication.

        - Mathematical representation of qubits
        
            Fundamentally, quantum computing uses linear algebra as the mathematical framework. A qubit's state is represented by a column vector with two elements. The two fundamental states of qubits, analogous to a bit's "0" and "1", are $|0\rangle$ and $|1\rangle$:

            $|0\rangle = \left(\begin{array}{c} 1 \\ 0 \end{array} \right) $, 
            $|1\rangle = \left(\begin{array}{c} 0 \\ 1 \end{array} \right) $.

            In the above equation, states are represented using Dirac Notation. $|\cdot \rangle$ is called a "ket". For each ket, there is a corresponding "bra" $\langle \cdot|$ representing the conjugate transpose of the ket.


        - Superposition

            As mentioned before, qubis can also exist in superposition states, i.e., different linear combinations of basis states:

            $|\psi\rangle = \alpha |0 \rangle + \beta |1 \rangle = \left(\begin{array}{c} \alpha \\ \beta \end{array} \right)$,

            where $\alpha$ and $\beta$ are complex numbers such that $\alpha\alpha^* + \beta\beta^* = 1 $, * denotes the complex conjugate.

            These $\alpha$ and $\beta$ are called amplitudes or probability amplitudes.

        - Measurement and probabilities of outcomes

            When we measure a qubit, we can not directly observe its superposition state. Instead, we observe the qubit collapsing into either state $|0\rangle$ or state $|1\rangle$. Namely, measurement in quantum computing is probabilistic. The amplitudes $\alpha$ and $\beta$ determine the probability of each of these outcomes. Specifically, for $|\psi\rangle$, the probability of measuring and observing $|0\rangle$ is equal to $|\alpha|^2$; the probability of measuring and observing $|1\rangle$ is equal to $|\beta|^2$. $|\cdot|^2$ denotes the multiplication of a number by its complex conjugate, i.e., $|\alpha|^2 = \alpha\alpha^*$. Since there are only two possible outcomes for one qubit, these probabilities must sum to 1, which explains why quantum states must be normalised.

        - Inner product

            The inner product between two quantum states is computed by taking the dot product of the statevectors. Apart from $|\psi\rangle$ (Equation 2), suppose we have another qubit state $|\phi\rangle$:

            $|\phi\rangle = \gamma |0 \rangle + \delta |1 \rangle $.

            The inner product between $|\psi\rangle$ and $|\phi\rangle$ is:

            $\langle \psi|\phi \rangle = \left(\begin{array}{cc} \gamma^* & \delta^* \end{array} \right) \cdot \left(\begin{array}{c} \alpha \\ \beta \end{array} \right) = 1 \cdot 0 + 0 \cdot 1 = 0$

    - Quantum circuits

        Quantum circuits are the fundamental framework for designing and implementing quantum algorithms.
    
        - Wires and registers

            A quantum circuit begins with a set of wires, each corresponding to a specific qubit in the quantum system. These wires provide a visual representation of the quantum information flow from the initial state, through various quantum gates, to the final measurement. 
            
            The qubits are arranged in sequence from top to bottom and are usually labeled numerically in this order, starting from 0, which is consistent with most quantum programming frameworks. A collection of qubits used together for quantum computations is referred to as a quantum register. The state of a quantum register is described by the combined quantum state of its qubits, which can represent a complex superposition of all possible classical states. For instance, a quantum register with $n$ qubits can represent $2^n$ possible states simultaneously, allowing for the representation and manipulation of a vast amount of information.

        - Quantum gates

            Quantum gates are a type of quantum operations that act on qubits. They are represented by unitary matrices and used to manipulate the state of qubits. An $ n \times n$ complex-valued matrix U is unitary if $UU^{\dagger} = U^{\dagger}U = I_n$, where $I_n$ is the $n$-dimensional identity matrix.

            - Single qubit gates

                - Pauli Gates

                X Gate: it flips the state of a qubit.

                $X = \left(\begin{array}{cc} 0 & 1 \\ 1 & 0 \end{array} \right)$

                Y Gate: it flips the state and the phase of a qubit.
                
                $Y = \left(\begin{array}{cc} 0 & -i \\ i & 0 \end{array} \right)$

                Z Gate: it adds a phase of $\pi$ to the state $|1\rangle$ component of a qubit.
                
                $Z = \left(\begin{array}{cc} 1 & 0 \\ 0 & -1 \end{array} \right)$

            - Phase Gate (S Gate): it adds a phase of $\frac{\pi}{2}$ to the state $|1\rangle$ component of a qubit.

                $S = \left(\begin{array}{cc} 1 & 0 \\ 0 & i \end{array} \right)$

            - Hadamard Gate (H Gate): it creates equal superpositions of $|0\rangle$ and $|0\rangle$ states.

                $H = \frac{1}{\sqrt{2}} \left(\begin{array}{cc} 1 & 1 \\ 1 & -1 \end{array} \right)$

            - Rotation Gates

                - RX Gate: it rotates a qubit around the x-axis of the Bloch sphere by an angle $\theta$.

                $ R_x(\theta) = \left(\begin{array}{cc} \cos\frac{\theta}{2} & -i\sin\frac{\theta}{2} \\ -i\sin\frac{\theta}{2} & \cos\frac{\theta}{2} \end{array} \right)$

                - RY Gate: it rotates a qubit around the y-axis of the Bloch sphere by an angle $\theta$.

                $ R_y(\theta) = \left(\begin{array}{cc} \cos\frac{\theta}{2} & -\sin\frac{\theta}{2} \\ \sin\frac{\theta}{2} & \cos\frac{\theta}{2} \end{array} \right)$

                - RZ Gate: it rotates a qubit around the z-axis of the Bloch sphere by an angle $\theta$.

                $ R_z(\theta) = \left(\begin{array}{cc} e^{-i\frac{\theta}{2}} & 0 \\ 0 & e^{i\frac{\theta}{2}} \end{array} \right)$

                - General Rotation Gate: it combines rotations around all three axes.

                $ R(\theta,\phi, \lambda) = e^{-i\frac{\phi}{2}} \left(\begin{array}{cc} \cos\frac{\theta}{2} & -e^{i\lambda}\sin\frac{\theta}{2} \\ e^{i\lambda}\sin\frac{\theta}{2} & e^{i(\phi + \lambda)}\cos\frac{\theta}{2} \end{array} \right)$

        - Measurements

            Another important operation is measurement, the final step of quantum computation. A measurement is represented by a meter-like symbol, which can be implemented on one or more qubits.

            - Projective measurements (von Neumann measurement)

                Projective measurements involves projecting the quantum state onto an eigenbasis of an observable. This gives us information about how much the basis vector contribute to a given state. Given a quantum state $|\psi\rangle$ the probability of observing it in state $|\phi\rangle$ if measuring it with respect to a basis that includes $|\phi\rangle$ is equal to $|\langle\phi|\psi\rangle|^2$.

            - Positive Operator-Valued Measure (POVM) 

                Unlike projective measurements, which are limited to orthogonal projections onto the eigenstates of an observable, POVMs allow for a broader range of measurements. This measurement uses a set of positive semi-definite operators $\set{E_i}$. Semi-definite means for $\langle \psi|E_i|\psi\rangle$ all quantum state $|\psi\rangle$. The sum of all POVM elements must satisfy $\sum_i E_i =I$. The probability of outcome i is $Tr(\rho E_i)$, where the trace of the product of the state $\rho$ and the POVM element $E_i$.

                POVMs encompass projective measurements as a special case. In projective measurements, the POVM elements are projections onto orthogonal subspaces corresponding to the eigenstates of an observable. However, POVMs are more general, as the need not be orthogonal or even projectors.

        - Observable

            In quantum computing and quantum mechanics, an observable is a physical quantity that can be measured. Observables are represented by Hermitian (or self-adjoint) operators acting on the state space of the quantum system. A Hermitian operator $\hat{O}$ has the property that $\hat{O} ={\hat{O}}^\dagger$, and therefore its eigenvalues, which represent possible measurement outcomes, are real numbers. For example, Pauli Z operation is Hermitian, and has eigenvalues 1 and -1 with corresponding eigenstates $|0\rangle$ and $|1\rangle$.

            Measuring an observable in a quantum system involves projecting the quantum state onto the eigenstates of the corresponding operator. The result of the measurement is one of the eigenvalues, and the quantum state collapses to the corresponding eigenstate. The probability of obtaining a particular eigenvalue is given by the square of the amplitude of the state's projection onto the corresponding eigenstate.

            If a quantum system is in a state $|\psi\rangle$, the expectation value of an observable $\hat{O}$ is $\langle \hat{O} \rangle = \langle \psi|\hat{O}|\psi\rangle$. This value represents the weighted average of measurements on a large number of identically prepared systems in state $|\psi\rangle$.

- Basic Quantum algorithms

    - Hadamard test

        The Hadamard test is a fundamental quantum algorithm used to estimate the expectation value of an operator with respect to a quantum state. It involves an ancillary qubit and a series of quantum gates applied to both the ancillary qubit and the target quantum state. 

        $|0\rangle |\psi\rangle \mapsto \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)|\psi\rangle \mapsto \frac{1}{\sqrt{2}}(|0\rangle |\psi\rangle + |1\rangle U |\psi\rangle) \mapsto \frac{1}{2}((|0\rangle + |1\rangle)|\psi\rangle + (|0\rangle - |1\rangle) U |\psi\rangle) = \frac{1}{2}(|0\rangle(|\psi\rangle + U |\psi\rangle) + |1\rangle(|\psi\rangle - U |\psi\rangle)) $

        Then measure the ancillary qubit in the computational basis. The probability of measuring $|0\rangle$ is $P(0) = \frac{1}{4}(\langle\psi|\psi\rangle + \langle\psi|U^{\dagger}|\psi\rangle + \langle\psi|U|\psi\rangle + \langle\psi|U^{\dagger}U|\psi\rangle) = \frac{1}{2}(1 + Re(\langle\psi|U|\psi\rangle))$. Similarly, the probability of measuring $|1\rangle$ is $P(1) = \frac{1}{2}(1 - Re(\langle\psi|U|\psi\rangle))$. We can estimate the real part of the expectation value of $U$ with respect to $|\psi\rangle$ is equal to $P(0)-P(1)$.


    - SWAP test

        The SWAP test is a quantum algorithm used to determine the similarity (or overlap) between two quantum states. Specifically, it can estimate the squared overlap $|\langle\phi|\psi\rangle|^2$ between two quantum states $|\phi\rangle$ and $|\psi\rangle$. The SWAP test involves one ancillary qubit and two quantum registers containing the states $|\phi\rangle$ and $|\psi\rangle$.


        $|0\rangle|\phi\rangle|\psi\rangle \mapsto \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)|\phi\rangle|\psi\rangle \mapsto \frac{1}{\sqrt{2}}(|0\rangle |\phi\rangle |\psi\rangle + |1\rangle |\psi\rangle|\phi\rangle) \mapsto \frac{1}{2}((|0\rangle + |1\rangle)|\phi\rangle |\psi\rangle + (|0\rangle - |1\rangle)|\psi\rangle|\phi\rangle) \mapsto \frac{1}{2}(|0\rangle(|\phi\rangle |\psi\rangle + |\psi\rangle|\phi\rangle) + |1\rangle(|\phi\rangle |\psi\rangle - |\psi\rangle|\phi\rangle))$

        Then measure the ancillary qubit in the computational basis. The probability of measuring $|0\rangle$ is $P(0) = \frac{1}{2}(1 + |\langle\phi|\psi\rangle|^2)$ and measuring $|0\rangle$ $P(1) = \frac{1}{2}(1 - |\langle\phi|\psi\rangle|^2)$. The squared overlap can be estimated as $ |\langle\phi|\psi\rangle|^2 = 2P(0) - 1$.

    - Quantum embedding
        Quantum embedding is to represent classical data in a quantum state space through certain quantum feature maps. 
        
        There are a few notions to be explained. 
        
        First of all, a quantum state space (also referred to as a Hilbert space) is a complex vector space equipped with an inner product. In this space, the norm of a vector (quantum state) must be 1, indicating it is a unit vector. The dimensionality of the Hilbert space corresponds to the number of independent quantum states that can exist within the system. For a system of $n$ qubits, the state space has a dimension of $2^n$. 
        
        Another notion is the quantum feature map. This process is analogous to the feature mapping in classical machine learning, where data is transformed into a higher-dimensional space, i.e. a feature space, to make it easier to identify patterns or features. It encodes a classical data point into a set of gate parameters within a quantum circuit, thereby generating a quantum state.

        Consider there are classical input data containing $M$ elements, with $N$ features each, $D = x^{(1)},..., x^{(m)},...,x^{(M)}$, where $x^{(m)}$ is a $N$-dimensional vector for $m = 1,…, M$. To embed this data into $n$ quantum subsystems ($n$ qubits or $n$ qumodes for discrete- and continuous-variable quantum computing, respectively), some of embedding techniques are elaborated below.

        - Basis embedding

            Basis embedding maps each input to a specific computational basis state of a qubit system. The input data should be expressed in the form of N-bit binary strings: $x^{(m)} = (b_1,.., b_N)$, where $b_i \in {0,1}$ for $i = 1,...,N$. Assuming all features are represented with unit binary precision (one bit), each input example $x^{(m)}$ can be directly mapped to the quantum state $| x^{(m)} \rangle$. Consequently, the number of quantum subsystems, $ n$, must be at least equal to $N$. The entire dataset can be represented as superpositions of computational basis states with each weighted by $\frac{1}{\sqrt{M}}$, as follows: 

            
            $|D\rangle = \frac{1}{\sqrt{M}} \sum_{m=1}^{M} |x(m)\rangle $

            For $N$ bits, there are $2^N$ possible basis states. Given $M \ll 2^N$, the basis embedding of $D$ will be sparse.

        - Amplitude embedding

            Amplitude embedding leverages the amplitudes of quantum states to represent the data.

            Given a normalised classical data vector $ \mathbf{x} = [x_1, x_2, \ldots, x_N]$ such that $\sum_{i=1}^{N} |x_i|^2 = 1$, amplitude embedding maps this vector into the amplitudes of a quantum state. The quantum state $|\psi\rangle$ can be written as:

            $ |\psi\rangle = \sum_{i=1}^{N} x_i |i\rangle $

            where $|i\rangle$ are the computational basis states of an $n$-qubit system (with $N = 2^n$), and the $x_i$ are the amplitudes associated with each basis state.

            Consider the classical dataset $D$ mentioned above. Its amplitude embedding can be easily understood by concatenating all the input $x^{(m)}$ together into one vector:

            $
            \alpha = C_{norm} [x^{(1)}_1, \ldots, x^{(1)}_N, x^{(2)}_1, \ldots, x^{(2)}_N, \ldots, x^{(M)}_1, \ldots, x^{(M)}_N]
            $

            where $C_{norm}$ is the normalization constant; this vector must be normalized such that $|\alpha|^2 = 1$. The input dataset can now be represented in the computational basis as:

            $
            |D\rangle = \sum_{i=1}^{2^n} \alpha_i |i\rangle
            $

            where $\alpha_i$ are the elements of the amplitude vector $\alpha$ and $|i\rangle$ are the computational basis states. The number of amplitudes to be encoded is $N \times M$. Since a system of $n$ qubits provides $2^n$ amplitudes, amplitude embedding requires $n \geq \log_2(NM)$ qubits. If $N \times M$ is less than $2^n$, non-informative constants can be padded to $\alpha$.
            
            Amplitude embedding allows for the representation of $N$ classical data points using $\log_2(N)$ qubits, providing an exponential reduction in the required number of qubits compared to directly encoding each data point as a separate qubit.

        There are various other embedding techniques with similar protocols. For instance, squeezing embedding and displacement embedding are employed in continuous-variable quantum computing models, where classical information is encoded in the parameters of the squeezing and displacement operators. Hamiltonian embedding, on the other hand, utilises an implicit method by encoding information through the evolution of a quantum system. This approach determines the time evolution of a quantum system, and by designing specific Hamiltonians, classical information can be encoded into the quantum state through its dynamic evolution.
        
        By encoding data into quantum states, quantum algorithms can exploit quantum parallelism, performing computations on all possible states simultaneously. This can lead to significant speedups for certain tasks, such as searching or optimisation.

### Variational classifier

```python
# Import PennyLane, the PennyLane-provided version of NumPy, and an optimizer
import pennylane as qml
from pennylane import numpy as np
from pennylane.optimize import NesterovMomentumOptimizer
```

```python
# create a quantum device that will run the circuits.
dev = qml.device("default.qubit")
```

#### Method 1: Basis Encoding

##### Quantum and classical nodes

```python
# Define the 4-qubit circuit's variational layer
def layer(layer_weights):
    # Apply a rotation on each wire and encode each wire's weight in the rotation angles
    for wire in range(4):
        qml.Rot(*layer_weights[wire], wires=wire)
    # Apply CNOT gates across the nearest neighbors
    for wires in ([0, 1], [1, 2], [2, 3], [3, 0]):
        qml.CNOT(wires)
```

```python
# Define the state preparation routine
def state_preparation(x):
    # Encode the input x into the quantum state (all entries of x are in {0, 1} and x has length 4)
    qml.BasisState(x, wires=[0, 1, 2, 3])
```

```python
# Define the quantum circuit that starts from the state preparation and followed by the variational layer
@qml.qnode(dev)
def circuit(weights, x):
    state_preparation(x)

    for layer_weights in weights:
        layer(layer_weights)

    # Return the expectation value of the Pauli-Z operator on the first wire
    return qml.expval(qml.PauliZ(0))
```

```python
# Define the full variational classifier model as a sum of the output of the quantum circuit, plus the trainable bias
def variational_classifier(weights, bias, x):
    return circuit(weights, x) + bias
```

##### Cost

```python
# Define the cost function using the standard square loss that measures the distance between target labels and model predictions
def square_loss(labels, predictions):
    # Use a call to qml.math.stack to allow subtracting the arrays directly
    return np.mean((labels - qml.math.stack(predictions)) ** 2)
```

```python
# Define the accuracy function that measures the percentage of correct predictions 
def accuracy(labels, predictions):
    acc = sum(abs(l - p) < 1e-5 for l, p in zip(labels, predictions))
    acc = acc / len(labels)
    return acc
```

```python
# Define the cost function that evaluates the square loss, which considers the features and labels in the iteration of the optimization routine
def cost(weights, bias, X, Y):
    predictions = [variational_classifier(weights, bias, x) for x in X]
    return square_loss(Y, predictions)
```

##### Optimization

```python
# Load the training data
data = np.loadtxt("data/parity_train.txt", dtype=int)
X = np.array(data[:, :-1])
Y = np.array(data[:, -1])
Y = Y * 2 - 1  # shift label from {0, 1} to {-1, 1}

for x,y in zip(X, Y):
    print(f"x = {x}, y = {y}")
```

```python
# Initialize the weights and bias
np.random.seed(0)
num_qubits = 4
num_layers = 2
weights_init = 0.01 * np.random.randn(num_layers, num_qubits, 3, requires_grad=True)
bias_init = np.array(0.0, requires_grad=True)

print("Weights:", weights_init)
print("Bias: ", bias_init)
```

```python
# Create an optimizer and set the batch size
opt = NesterovMomentumOptimizer(0.5)
batch_size = 5
```

```python
# Run the optimization routine and track the accuracy and cost
weights = weights_init
bias = bias_init
for it in range(100):

    # Update the weights by one optimizer step, using only a limited batch of data
    batch_index = np.random.randint(0, len(X), (batch_size,))
    X_batch = X[batch_index]
    Y_batch = Y[batch_index]
    weights, bias = opt.step(cost, weights, bias, X=X_batch, Y=Y_batch)

    # Compute the outputs of the variational classifier and turn them into predictions in $\{-1,1\}$ by taking the sign of the output.
    predictions = [np.sign(variational_classifier(weights, bias, x)) for x in X]
    # Compute the cost and accuracy of the current predictions
    current_cost = cost(weights, bias, X, Y)
    acc = accuracy(Y, predictions)

    print(f"Iter: {it+1:4d} | Cost: {current_cost:0.7f} | Accuracy: {acc:0.7f}")
```

```python
# Load the test data
data = np.loadtxt("data/parity_test.txt", dtype=int)
X_test = np.array(data[:, :-1])
Y_test = np.array(data[:, -1])
Y_test = Y_test * 2 - 1  # shift label from {0, 1} to {-1, 1}

# Compute the predictions on the test set  
predictions_test = [np.sign(variational_classifier(weights, bias, x)) for x in X_test]

for x,y,p in zip(X_test, Y_test, predictions_test):
    print(f"x = {x}, y = {y}, pred={p}")

# Compute the accuracy on the test set    
acc_test = accuracy(Y_test, predictions_test)
print("Accuracy on unseen data:", acc_test)
```

#### Method 2: Amplitude Encoding

##### Quantum and classical nodes

```python
# Transform the input data to the angles of the quantum state preparation circuit
def get_angles(x):
    beta0 = 2 * np.arcsin(np.sqrt(x[1] ** 2) / np.sqrt(x[0] ** 2 + x[1] ** 2 + 1e-12))
    beta1 = 2 * np.arcsin(np.sqrt(x[3] ** 2) / np.sqrt(x[2] ** 2 + x[3] ** 2 + 1e-12))
    beta2 = 2 * np.arcsin(np.linalg.norm(x[2:]) / np.linalg.norm(x))

    return np.array([beta2, -beta1 / 2, beta1 / 2, -beta0 / 2, beta0 / 2])

# Define the state preparation routine
def state_preparation(a):
    qml.RY(a[0], wires=0)

    qml.CNOT(wires=[0, 1])
    qml.RY(a[1], wires=1)
    qml.CNOT(wires=[0, 1])
    qml.RY(a[2], wires=1)

    qml.PauliX(wires=0)
    qml.CNOT(wires=[0, 1])
    qml.RY(a[3], wires=1)
    qml.CNOT(wires=[0, 1])
    qml.RY(a[4], wires=1)
    qml.PauliX(wires=0)
```

```python
# Test the state preparation routine

x = np.array([0.53896774, 0.79503606, 0.27826503, 0.0], requires_grad=False)
ang = get_angles(x)


@qml.qnode(dev)
def test(angles):
    state_preparation(angles)

    return qml.state()


state = test(ang)

print("x               : ", np.round(x, 6))
print("angles          : ", np.round(ang, 6))
print("amplitude vector: ", np.round(np.real(state), 6))
```

```python
# Update the variational layer function
def layer(layer_weights):
    for wire in range(2):
        qml.Rot(*layer_weights[wire], wires=wire)
    qml.CNOT(wires=[0, 1])

# Redefine the cost function 
def cost(weights, bias, X, Y):
    # Transpose the batch of input data in order to make the indexing
    # in state_preparation work
    predictions = variational_classifier(weights, bias, X.T)
    return square_loss(Y, predictions)
```

##### Data processing

```python
# Load the iris data set
data = np.loadtxt("data/iris_classes1and2_scaled.txt")
X = data[:, 0:2]
print(f"First X sample (original)  : {X[0]}")

# Pad the vectors to size 2^2=4 with constant values
padding = np.ones((len(X), 2)) * 0.1
X_pad = np.c_[X, padding]
print(f"First X sample (padded)    : {X_pad[0]}")

# Normalize each input
normalization = np.sqrt(np.sum(X_pad**2, -1))
X_norm = (X_pad.T / normalization).T
print(f"First X sample (normalized): {X_norm[0]}")

# Create the features which are the angles for state preparation
features = np.array([get_angles(x) for x in X_norm], requires_grad=False)
print(f"First features sample      : {features[0]}")

Y = data[:, -1]
```

```python
# Split the data into training and validation sets
np.random.seed(0)
num_data = len(Y)
num_train = int(0.75 * num_data)
index = np.random.permutation(range(num_data))
feats_train = features[index[:num_train]]
Y_train = Y[index[:num_train]]
feats_val = features[index[num_train:]]
Y_val = Y[index[num_train:]]

# We need these later for plotting
X_train = X[index[:num_train]]
X_val = X[index[num_train:]]
```

##### Optimization

##### Build a 4-qubit quantum circuit on a PennyLane `default.qubit` device consisting of:
1.  an embedding layer of local $R_y$ rotations (with angles scaled by a
    factor of $\pi$);
2.  a random circuit of `n_layers`;
3.  a final measurement in the computational basis, estimating $4$
    expectation values.

```python
# Define the quantum device
dev = qml.device("default.qubit", wires=4)

# Random circuit parameters
rand_params = np.random.uniform(high=2 * np.pi, size=(n_layers, 4))

# Define the quantum circuit
@qml.qnode(dev)
def circuit(phi):
    # Encoding of 4 classical input values
    for j in range(4):
        qml.RY(np.pi * phi[j], wires=j)

    # Random quantum circuit
    RandomLayers(rand_params, wires=list(range(4)))

    # Measurement producing 4 classical output values
    return [qml.expval(qml.PauliZ(j)) for j in range(4)]
```

##### Implement the convolution scheme:
1.  the image is divided into squares of $2 \times 2$ pixels;
2.  each square is processed by the quantum circuit;
3.  the $4$ expectation values are mapped into $4$ different channels of
    a single output pixel.

```python
def quanv(image):
    """Convolves the input image with many applications of the same quantum circuit."""
    out = np.zeros((14, 14, 4))

    # Loop over the coordinates of the top-left pixel of 2X2 squares
    for j in range(0, 28, 2):
        for k in range(0, 28, 2):
            # Process a squared 2x2 region of the image with a quantum circuit
            q_results = circuit(
                [
                    image[j, k, 0],
                    image[j, k + 1, 0],
                    image[j + 1, k, 0],
                    image[j + 1, k + 1, 0]
                ]
            )
            # Assign expectation values to different channels of the output pixel (j/2, k/2)
            for c in range(4):
                out[j // 2, k // 2, c] = q_results[c]
    return out
```

##### Apply quantum pre-processing layer to all the images of the dataset

```python
if PREPROCESS == True:
    q_train_images = []
    print("Quantum pre-processing of train images:")
    # Process all training images
    for idx, img in enumerate(train_images): 
        print("{}/{}        ".format(idx + 1, n_train), end="\r")
        # Apply the quantum convolution
        q_train_images.append(quanv(img))
    # Convert list to array
    q_train_images = np.asarray(q_train_images)

    # Process all test images
    q_test_images = []
    print("\nQuantum pre-processing of test images:")
    for idx, img in enumerate(test_images):
        print("{}/{}        ".format(idx + 1, n_test), end="\r")
        # Apply the quantum convolution
        q_test_images.append(quanv(img))
    # Convert list to array
    q_test_images = np.asarray(q_test_images)

    # Save pre-processed images
    np.save(SAVE_PATH + "q_train_images.npy", q_train_images)
    np.save(SAVE_PATH + "q_test_images.npy", q_test_images)


# Load pre-processed images
q_train_images = np.load(SAVE_PATH + "q_train_images.npy")
q_test_images = np.load(SAVE_PATH + "q_test_images.npy")
```

#### **Task 4**

Learn the sine function on the interval [0, 2π]

```python
import jax
import jax.numpy as jnp
import jaxopt
```

#### Variational quantum algorithm (VQA)

##### Ansatz

Apply a $\mathbf{\theta}$-parameterized Unitary operator U($\mathbf{\theta}$) to the input state 

$ |\psi_{out}\rangle(x_i , \mathbf{\theta}) = U(\mathbf{\theta})|\psi_{in}(x_i)\rangle$

$U(\theta^{(i)}_j)=R^X_j(\theta^{(i)}_{j1})R^Z_j(\theta^{(i)}_{j2})R^X_j(\theta^{(i)}_{j3})$

```python
n_wires = 3 # number of qubits
depth = 3 # number of layers
data = np.linspace(0, 2*np.pi, 30) # input data
target = np.sin(data) # target data

# initialise the parameters
params = np.random.uniform(0, 2 * np.pi, n_wires*3)
params = list(params)
```

```python
dev = qml.device("default.qubit", wires=n_wires)
@qml.qnode(dev, interface="jax", diff_method="parameter-shift")
def circuit(data, params):
    """Quantum circuit ansatz"""

    # data embedding
    for i in range(n_wires):
        qml.RY(data, wires=i)

    # trainable ansatz
    for i in range(depth):
        for j in range(n_wires):
            qml.CNOT(wires=[j, (j + 1) % n_wires])

        for j in range(n_wires):
            qml.RX(params[3*j+0], wires=j)
            qml.RZ(params[3*j+1], wires=j)
            qml.RX(params[3*j+2], wires=j)
        

    # we use a sum of local Z's as an observable since a
    # local Z would only be affected by params on that qubit.
    
    return qml.expval(qml.PauliZ(0))
```

```python
qml.draw_mpl(circuit)(data, params)
```

##### Cost function 

$C(\mathbf{\theta}) = \sum\limits_i (\sin x - u(x))^2$

where $u(x)$ is the approximated function

```python
@jax.jit
def cost(params):
    """Cost function to be minimized"""

    # Compute prediction
    preds = [circuit(data[i], params) for i in range(len(data))]

    # Compute mean squared error
    return jnp.mean((jnp.array(preds) - target) ** 2)
```

##### Optimizer

```python
# initialize the optimizer
opt = jaxopt.GradientDescent(cost, stepsize=0.01, tol = 1e-9)
# store the parameters
param_store = {'1':params}
loss_store = {'1':cost(params)}
n_steps = 1000
```

```python
opt_state = opt.init_state(params)
param_store = {'1':params}
loss_store = {'1':cost(params)}

for i in range(1, n_steps+1):
    params, opt_state = opt.update(params, opt_state)
    param_store.update({str(i+1):params})
    loss_store.update({str(i+1):cost(params)})

    if (i+1)%10 == 0: 
        print(f"Step = {(i+1)},  Cost function = {loss_store[str(i+1)]:.8f} ")

    if loss_store[str((i+1))]-loss_store[str((i))] < 1e-6:
        print(f"Step = {(i+1)},  Cost function = {loss_store[str((i+1))]:.8f} ")
        break
```

##### Result

```python
approx = circuit(data, params)

plt.plot(data, target, 'bo', label = 'Exact')
plt.plot(data, approx, 'r-', label = 'Approximated')
```
